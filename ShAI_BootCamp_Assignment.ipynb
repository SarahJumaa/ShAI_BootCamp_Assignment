```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

```python
#########TASK 1############
rows, columns = df.shape
print(f"عدد الصفوف: {rows}, عدد الأعمدة: {columns}")
data_types = df.dtypes
print("أنواع البيانات:")
print(data_types)
missing_values = df.isnull().sum()
print("القيم المفقودة في كل عمود:")
print(missing_values)
```

```python
#########TASK 2############
average_salary = df['TotalPay'].mean()
print(f" المتوسط: {average_salary}")
median_salary = df['TotalPay'].median()
print(f" الوسيط: {median_salary}")
mode_salary = df['TotalPay'].mode()
print(f" الوضع: {mode_salary}")
min_salary = df['TotalPay'].min()
print(f" الحد الأدنى: {min_salary}")
max_salary = df['TotalPay'].max()
print(f" الحد الأعلى: {max_salary}")
salary_range = max_salary - min_salary
print(f" نطاق الرواتب: {salary_range}")
standard_deviation_salary = df['TotalPay'].std()
print(f" الانحراف المعياري: {standard_deviation_salary}")
```

```python
#########TASK 3############
data_cleaned = df.dropna()
print(f"عدد الصفوف: {rows}")
```

```python
#########TASK 3############
mean_basepay = df['BasePay'].mean()
df['BasePay'].fillna(mean_basepay, inplace=True)
mean_overtimepay = df['OvertimePay'].mean()
df['OvertimePay'].fillna(mean_overtimepay, inplace=True)
std_otherpay = df['OtherPay'].std()
df['OtherPay'].fillna(std_otherpay, inplace=True)
mode_benefits = df['Benefits'].mode()
df['Benefits'].fillna(mode_benefits[0], inplace=True)
```

```python
#########TASK 4############
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.distplot(df['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay (Salary)')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10, 10))
agency_distribution = df['JobTitle'].value_counts()
plt.pie(agency_distribution, labels=agency_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Employees by Agency')
plt.show()
```

```python
#########TASK 5############
average_salary_by_job_title = df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False)

print("Top 10 Job Titles with Highest Average Salary:")
print(average_salary_by_job_title.head(10))
print("\n")

average_salary_by_agency = df.groupby('Agency')['TotalPay'].mean().sort_values(ascending=False)

print("Top 10 Agencies with Highest Average Salary:")
print(average_salary_by_agency.head(10))
```

```python
#########TASK 6############
average_salary_by_job_year = df.groupby(['JobTitle', 'Year'])['TotalPay'].mean().reset_index()

for year in df['Year'].unique():
    print(f"Top 10 Job Titles with Highest Average Salary in {year}:")
    top_job_titles = average_salary_by_job_year[average_salary_by_job_year['Year'] == year]
    top_job_titles = top_job_titles.sort_values(by='TotalPay', ascending=False).head(10)
    print(top_job_titles)
    print("\n")


average_salary_by_agency_year = df.groupby(['Agency', 'Year'])['TotalPay'].mean().reset_index()


for year in df['Year'].unique():
    print(f"Top 10 Agencies with Highest Average Salary in {year}:")
    top_agencies = average_salary_by_agency_year[average_salary_by_agency_year['Year'] == year]
    top_agencies = top_agencies.sort_values(by='TotalPay', ascending=False).head(10)
    print(top_agencies)
    print("\n")
```
